# Brand

```python
df = df.dropna(subset = ['department', 'first_date'])
```

```python
import re

def extract_brand(text):
    result = re.search(r'\bVisit\b\s+\bthe\b\s+(.+?)\s+\bStore\b', text)    
    if result:
        brand = result.group(1)
        return brand.capitalize()  
    result = re.search(r'Brand:\s*(.+)', text)
    if result:
        brand = result.group(1).strip().lower()
        return brand.capitalize() 
    return None
```

```python
df['brand'] = df['brand'].apply(extract_brand)
```

```python
def convert_to_basic_color(color_name):
    color_keywords = [
        'Red', 'Green', 'Blue', 'Yellow', 'Pink', 
        'Purple', 'Black', 'White', 'Brown', 'Grey', 'Orange', 'Navy'
    ]

    color_variants = {
        'Red': ['Crimson', 'Scarlet', 'Burgundy', 'Cherry', 'Tomato'],
        'Green': ['Olive', 'Emerald', 'Mint', 'Lime', 'Forest'],
        'Blue': ['Navy', 'Sky Blue', 'Cobalt', 'Turquoise', 'Teal'],
        'Yellow': ['Golden', 'Mustard', 'Lemon', 'Canary', 'Honey'],
        'Pink': ['Rose', 'Fuchsia', 'Blush', 'Magenta', 'Salmon'],
        'Purple': ['Lavender', 'Violet', 'Mauve', 'Plum', 'Indigo'],
        'Black': ['Jet Black', 'Charcoal', 'Ebony', 'Onyx', 'Coal'],
        'White': ['Ivory', 'Snow', 'Cream', 'Pearl', 'Alabaster'],
        'Brown': ['Chocolate', 'Tan', 'Beige', 'Café', 'Mahogany'],
        'Grey': ['Slate', 'Ash', 'Charcoal', 'Silver', 'Steel'],
        'Orange': ['Tangerine', 'Peach', 'Coral', 'Burnt Orange', 'Amber'],
        'Navy': ['Midnight Blue', 'Prussian Blue', 'Cobalt Blue', 'Indigo', 'Royal Blue']
    }

    color_keywords_lower = [i.lower() for i in color_keywords]
    color_name = color_name.lower()
    delimiters = [' ', '/', '-']

    for deli in delimiters:
        for i in color_name.split(deli):
            if i in color_keywords_lower:
                return i.capitalize()

    for basic_color, variants in color_variants.items():
        for variant in variants:
            if variant.lower() in color_name:
                return basic_color
    
    return None
```

```python
df.loc[:, 'color'] = df['color'].apply(convert_to_basic_color)
df
```

# Fabric

```python
import re

def parse_fabric(s):
    matches = re.findall(r'(\d+)%\s*(\w+)', s)
    if not matches:  
        return None
    fabric_dict = {fabric: int(proportion)/100 for proportion, fabric in matches}
    return fabric_dict
```

```python
df['fabric'] = df['fabric'].apply(parse_fabric)
```

```python
df = df.dropna(subset =['fabric'])
df.reset_index(drop=True, inplace=True)
```

# Care

```python
df.head()
```

```python
def process_care(instruction):
    base_care = ['Machine Wash', 'Hand Wash', "Bleach", "Tumble Dry"]
    
    care_dict = {care: 0 for care in base_care}
    
    instruction = instruction.lower()
    
    for care in base_care:
        if care.lower() in instruction:
            care_dict[care] = 1
    if all(value == 0 for value in care_dict.values()):
        return None
    
    return care_dict
```

```python
df['care'] = df['care'].apply(process_care)
df.reset_index(drop=True)
```

```python
df = df.dropna(subset =['care'])
```

```python
df_care = pd.json_normalize(df['care'])
```

# Closure

```python
df.head(1)
```

```python
df['closure'].value_counts(dropna=False)
```

```python
df['closure'] = df['closure'].fillna('No closure')
df.replace('Pull-On','Pull on', inplace=True)
```

```python
def process_closure(txt):
    base_closures = ['Pull on', 'Tie','Zipper','Button', 'No closure','Elastic','Lace Up','Drawstring']
    
    closure_dict = {care: 0 for care in base_closures}
    
    txt = txt.lower()
    
    for closure in base_closures:
        if closure.lower() in txt:
            closure_dict[closure] = 1
    if all(value == 0 for value in closure_dict.values()):
        return None
    
    return closure_dict
```

```python
df['closure'] = df['closure'].apply(process_closure)
df = df.dropna(subset =['closure'])
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df_closure = pd.json_normalize(df['closure'])
```

```python
def month_to_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'
```

```python
df['season'] = df['month'].apply(month_to_season)
```
