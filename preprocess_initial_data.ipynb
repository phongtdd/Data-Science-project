```python
import re

def extract_brand(text):
    result = re.search(r'\bVisit\b\s+\bthe\b\s+(.+?)\s+\bStore\b', text)    
    if result:
        brand = result.group(1)
        return brand.capitalize()  
    result = re.search(r'Brand:\s*(.+)', text)
    if result:
        brand = result.group(1).strip().lower()
        return brand.capitalize() 
    return None
```

```python
df['brand'] = df['brand'].apply(extract_brand)
```

# Color

```python
def convert_to_basic_color(color_name):
    color_keywords = [
        'Red', 'Green', 'Blue', 'Yellow', 'Pink', 
        'Purple', 'Black', 'White', 'Brown', 'Grey', 'Orange', 'Navy'
    ]

    color_variants = {
        'Red': ['Crimson', 'Scarlet', 'Burgundy', 'Cherry', 'Tomato', 'Rosewood', 'Firebrick', 'Ruby'],
        'Green': ['Olive', 'Emerald', 'Mint', 'Lime', 'Forest', 'Jade', 'Sage', 'Seafoam'],
        'Blue': ['Navy', 'Sky Blue', 'Cobalt', 'Turquoise', 'Teal', 'Azure', 'Royal Blue', 'Cerulean'],
        'Yellow': ['Golden', 'Mustard', 'Lemon', 'Canary', 'Honey', 'Butter', 'Amber', 'Daffodil'],
        'Pink': ['Rose', 'Fuchsia', 'Blush', 'Magenta', 'Salmon', 'Bubblegum', 'Hot Pink', 'Peony'],
        'Purple': ['Lavender', 'Violet', 'Mauve', 'Plum', 'Indigo', 'Lilac', 'Amethyst', 'Eggplant'],
        'Black': ['Jet Black', 'Charcoal', 'Ebony', 'Onyx', 'Coal', 'Raven', 'Sable', 'Midnight'],
        'White': ['Ivory', 'Snow', 'Cream', 'Pearl', 'Alabaster', 'Linen', 'Ghost White', 'Eggshell'],
        'Brown': ['Chocolate', 'Tan', 'Beige', 'Café', 'Mahogany', 'Coffee', 'Copper', 'Walnut'],
        'Grey': ['Slate', 'Ash', 'Charcoal', 'Silver', 'Steel', 'Graphite', 'Smoke', 'Zinc'],
        'Orange': ['Tangerine', 'Peach', 'Coral', 'Burnt Orange', 'Amber', 'Cantaloupe', 'Pumpkin', 'Apricot'],
        'Navy': ['Midnight Blue', 'Prussian Blue', 'Cobalt Blue', 'Indigo', 'Royal Blue', 'Sapphire', 'Blueberry', 'Steel Blue']
    }


    color_keywords_lower = [i.lower() for i in color_keywords]
    color_name = color_name.lower()
    delimiters = [' ', '/', '-']

    for deli in delimiters:
        for i in color_name.split(deli):
            if i in color_keywords_lower:
                return i.capitalize()

    for basic_color, variants in color_variants.items():
        for variant in variants:
            if variant.lower() in color_name:
                return basic_color
    
    return None
```

```python
df.loc[:, 'color'] = df['color'].apply(convert_to_basic_color)
```

# Fabric

```python
import re

material_variants = {
    "Polyester": [
        "Recycled Polyester", "Polyester and", "Polyester French Terry", "Polyester Blend", 
        "Polyester Sherpa", "Polyester Fleece", "Polyester Plush Fleece", "Polyester Shell", 
        "Polyester Thermalock WP", "Polyester Woven Shell Bonded to a", "Polyester Microfleece Lining", 
        "Recycled Polyester with Non", "Recycled Polyester Plain Weave with Non", "Pollyester", 
        "Polyeste r", "Poly", "Polyester"
    ],
    "Spandex": [
        "Elastane", "Spandex for", "Elasthan", "Spandex Great Material for Women's Sportswear", 
        "Elastane Ribbed Fabric", "Spandex Lining", "Spandex Butt Padded Shapewear", 
        "Spandex Tummy Control Shapewear Shorts", "Modal and Spandex", "Spandex or", "Spendex", "Span", 'Spandex'
    ],
    "Nylon": ["Nylon", "Polyamide", "Recycled Polyamide", "Polyamide and", "Nylon and"],
    "Cotton": [
        "Pima Cotton", "Cotton", "USA Cotton", "Combed Cotton", "Ring Spun Cotton", 
        "Climate Conscious Cotton", "Cotton Textured Jersey", "Cotton Better Cotton", 
        "Cotton US Cotton Trust Protocol Equivalent", "Combed Ringspun Cotton", "Recycled Cotton", 
        "Cotton Jersey", "Organic Cotton", "Fine Jersey Preshrunk Cotton", "Cotton Fabric"
    ],
    "Rayon": ["Rayon", "FSC Rayon", "Viscose", "Rayon Lining", "Viscose from Bamboo", "Rayon and", "Rayon Cotton"],
    "Leather": ["Leather", "Faux Leather", "Real Lambskin Leather"],
    "Lyocell": ["Tencel Lyocell", "Lyocell"],
    "Acrylic": ["Acrylic"],
    "Modal": ["Modal", "Modal and Spandex"],
    "Linen": ["Linen"],
    "Silk": ["Silk"],
    "Wool": ["Wool"]
}

drop_list = [
    "Dacron", "Terylene", "PU", "Polyurethane", "Chinlon", "PBT", "Flax", "Polycotton", 
    "Synthetic", "Blend Materials", "Fabric", "Knitted", "Chiffon", "Velvet", "Faux Suede", "Satin", ""
]


flattened_materials = {v.lower(): k for k, values in material_variants.items() for v in values}

def parse_fabric(s):
    matches = re.findall(r'(\d+(\.\d+)?)%\s*([a-zA-Z\s]+)', s)
    
    if not matches:
        return None
    
    fabric_dict = {}
    for proportion, _, fabric in matches:
        fabric = fabric.strip().lower()
        
        for drop in drop_list:
            if drop == fabric:
                break
        
        if fabric in flattened_materials:
            material_name = flattened_materials[fabric]
            fabric_dict[material_name] = float(proportion) / 100
    
    return fabric_dict if fabric_dict else None
```

```python
df['fabric'] = df['fabric'].apply(parse_fabric)
```

```python
df = df.dropna(subset =['fabric'])
df = df.reset_index(drop=True)
```

```python
df_expanded = pd.json_normalize(df['fabric'])
df_expanded = df_expanded.fillna(0)
```

```python
row_sums = df_expanded.sum(axis=1)
filter_index = row_sums[row_sums == 1].index
```

# Care

```python
def process_care(instruction):
    base_care = ['Machine Wash', 'Hand Wash', "Bleach", "Tumble Dry"]
    
    care_dict = {care: 0 for care in base_care}
    
    instruction = instruction.lower()
    
    for care in base_care:
        if care.lower() in instruction:
            care_dict[care] = 1
    if all(value == 0 for value in care_dict.values()):
        return None
    
    return care_dict
```

```python
df['care'] = df['care'].apply(process_care)
```

```python
df = df.dropna(subset =['care'])
df =df.reset_index(drop=True)
```

# Department

```python
# Function to fill name into department

def fill_depart(name, o):
    if pd.notna(o):
        return o
    else:
        departs = ['Women', 'Men', 'Unisex','Girls','Boys']
        for i in departs:
            if i.lower() in name.lower():
                return i
    return None
df['department'] = df.apply(lambda row: fill_depart(row['name'], row['department']), axis=1)
```

```python
df['department']= df['department'].replace({'womens': 'Women','mens':'Men', 'boys':"Men",'girls':"Women","Boys":'Men','Girls':'Women'})
df['department']= df['department'].str.replace(r'^unisex-\w+', 'Unisex', regex=True)
```

```python
df = df.dropna(subset='department')
df = df.reset_index(drop=True)
```

```python
departments = ['Women','Men','Unisex']
df = df[df['department'].isin(departments)]
```

```python
df['closure'] = df['closure'].fillna('No closure')
df.replace('Pull-On','Pull on', inplace=True)
```

```python
def process_closure(txt):
    base_closures = ['Pull on', 'Tie','Zipper','Button', 'No closure','Elastic','Lace Up','Drawstring']
    
    closure_dict = {care: 0 for care in base_closures}
    
    txt = txt.lower()
    
    for closure in base_closures:
        if closure.lower() in txt:
            closure_dict[closure] = 1
    if all(value == 0 for value in closure_dict.values()):
        return None
    
    return closure_dict
```

```python
df['closure'] = df['closure'].apply(process_closure)
df = df.dropna(subset =['closure'])
df = df.reset_index(drop=True)
```

```python
df_closure = pd.json_normalize(df['closure'])
```
