```python
import re
def parse_fabric(s):
    matches = re.findall(r'(\d+)%\s+(\w+)', s)
    fabric_dict = {fabric: int(proportion)/100 for proportion, fabric in matches}
    return fabric_dict
```

```python
df['fabric'] = df['fabric'].apply(parse_fabric)
```

```python
df_expanded = pd.json_normalize(df['fabric'])
```

```python
df1 = pd.concat([df.reset_index(drop=True), df_expanded.reset_index(drop=True)], axis=1)
```

```python
def month_to_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'
```
